### 7:1:2 비율

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

```

```python
import json
# jsonl 파일 읽어서 list에 저장
def jsonlload(fname, encoding="utf-8"):
    json_list = []
    with open(fname, encoding=encoding) as f:
        for line in f.readlines():
            json_list.append(json.loads(line))
    return json_list
df=jsonlload('total.jsonl')
df
```

```python
df=pd.DataFrame(df)
df
```

```python
df['label']
```

```python

train_set,dev_set = train_test_split(df, test_size=0.3, random_state=10, stratify=df['label'])
dev_set,test_set = train_test_split(dev_set,test_size=0.67,random_state=10,stratify=dev_set['label'])
```

```python
train_set=train_set.values.tolist()
dev_set=dev_set.values.tolist()
test_set=test_set.values.tolist()

train_set
```

```python
# jsonl 파일 읽어서 list에 저장
def jsonlload(fname, encoding="utf-8"):
    json_list = []
    with open(fname, encoding=encoding) as f:
        for line in f.readlines():
            json_list.append(json.loads(line))
    return json_list


def jsondump(j, fname):
    with open(fname, "w", encoding="UTF8") as f:
        for line in j:
            temp=dict()
            temp={'id':"",'category':"",'review_content':"",'label':""}
            temp['id']=line[0]
            temp['category']=line[1]
            temp['review_content']=line[2]
            temp['label']=line[3]
            json.dump(temp, f, ensure_ascii=False)
            f.write("\n")

jsondump(train_set,'train_data.jsonl')
jsondump(dev_set,'dev_data.jsonl')
jsondump(test_set,'test_data.jsonl')

```
